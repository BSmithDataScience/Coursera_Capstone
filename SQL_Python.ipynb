### Problem 8: List the top 5 Community Areas by average College Enrollments [number of students]

```python
query = "SELECT * FROM report_cards"
df = pd.read_sql(query,mydb)
df.head()
```

```python
query = "SELECT community_area_number as c_num, avg(college_enrollment) as avg_enroll\
    FROM report_cards GROUP BY community_area_number \
    ORDER BY avg_enroll DESC LIMIT 5"
show_sql(query)
```

### Problem 9: Use a sub-query to determine which Community Area has the least value for school Safety Score?

```python
query = "SELECT community_area_number,community_area_name FROM report_cards \
    WHERE safety_score = (SELECT min(safety_score) FROM report_cards)"
show_sql(query)
```

### Problem 10: [Without using an explicit JOIN operator] Find the Per Capita Income of the Community Area which has a school Safety Score of 1.

```python
query = "SELECT per_capita_income FROM census_data \
    WHERE community_area_number = (SELECT community_area_number FROM report_cards WHERE safety_score = 1)"
show_sql(query)
```
